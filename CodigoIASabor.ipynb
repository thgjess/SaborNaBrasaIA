```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor

df_excel = pd.read_excel('historico_vendas_trimestre_1.xlsx', sheet_name='Vendas Jan-Mar 2026')

mapa_dias = {'Segunda': 0, 'Terça': 1, 'Quarta': 2, 'Quinta': 3, 'Sexta': 4, 'Sábado': 5, 'Domingo': 6}
df_excel['Dia_Numerico'] = df_excel['Dia da Semana'].map(mapa_dias)

X = df_excel[['Dia_Numerico']]

modelo_marmita = RandomForestRegressor(n_estimators=100, random_state=42)
modelo_marmita.fit(X, df_excel['Marmita Executiva (Qtd)'])

modelo_mandioquinha = RandomForestRegressor(n_estimators=100, random_state=42)
modelo_mandioquinha.fit(X, df_excel['Caldo Mandioquinha (Qtd)'])

modelo_carne = RandomForestRegressor(n_estimators=100, random_state=42)
modelo_carne.fit(X, df_excel['Caldo Carne (Qtd)'])

print("Treinado")
```

```python
import pandas as pd

dias_proxima_semana = pd.DataFrame({'Dia_Numerico': [0, 1, 2, 3, 4, 5, 6]})
nomes_dias = ['Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta', 'Sábado', 'Domingo']

prev_marmita = modelo_marmita.predict(dias_proxima_semana)
prev_mandioquinha = modelo_mandioquinha.predict(dias_proxima_semana)
prev_carne = modelo_carne.predict(dias_proxima_semana)

df_previsoes = pd.DataFrame({
    'Dia da Semana': nomes_dias,
    'Marmita Executiva': prev_marmita.astype(int),
    'Caldo Mandioquinha': prev_mandioquinha.astype(int),
    'Caldo Carne': prev_carne.astype(int)
})

print("Previsões calculadas baseadas nos padrões de Jan-Mar:")
print(df_previsoes.to_string(index=False))
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(10, 6))

posicoes = np.arange(len(df_previsoes['Dia da Semana']))
largura = 0.25

plt.bar(posicoes - largura, df_previsoes['Marmita Executiva'], largura, label='Marmita Executiva', color='#4682B4')
plt.bar(posicoes, df_previsoes['Caldo Mandioquinha'], largura, label='Caldo Mandioquinha', color='#FFA500')
plt.bar(posicoes + largura, df_previsoes['Caldo Carne'], largura, label='Caldo Carne', color='#8B0000')

plt.xticks(posicoes, df_previsoes['Dia da Semana'])
plt.xlabel('Dia da Semana', fontsize=12)
plt.ylabel('Quantidade de Vendas Prevista', fontsize=12)
plt.title('Projeção de Demanda Baseada no 1º Trimestre (IA)', fontsize=14, fontweight='bold')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
